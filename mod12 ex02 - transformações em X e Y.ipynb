# Regressão 01 - tarefa 02 - transformações em X e Y

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

import statsmodels.formula.api as smf
%matplotlib inline
````

````python
tips = sns.load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
reg_pct_net = smf.ols('tip_pct ~ np.log(net_bill)', data = tips).fit()
print(reg_pct_net.summary())

sns.scatterplot(x = 'net_bill', y = reg_pct_net.resid, data = tips, alpha = .75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x = 'net_bill' , y = 'tip_pct', data = tips)
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
reg_pct_net2 = smf.ols('tip_pct ~ net_bill + np.power(net_bill,2)', data = tips).fit()
print(reg_pct_net2.summary())

sns.scatterplot(x = 'net_bill', y = reg_pct_net2.resid, data = tips, alpha = .75)
plt.axhline(y=0, color='r', linestyle='--')
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
reg_pct_net3 = smf.ols('np.log(tip_pct) ~ net_bill', data = tips).fit()
print(reg_pct_net3.summary())

sns.scatterplot(x = 'net_bill', y = reg_pct_net3.resid, data = tips, alpha = .75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
tips['pred_tip'] = np.exp(reg_pct_net3.fittedvalues)

tips[['pred_tip', 'tip_pct']].corr().iloc[0,1]**2
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
reg_pct_net4 = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data = tips).fit()
print(reg_pct_net4.summary())

sns.scatterplot(x = 'net_bill', y = reg_pct_net4.resid, data = tips, alpha = .75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
tips['pred_tip'] = np.exp(reg_pct_net4.fittedvalues)

tips[['pred_tip', 'tip_pct']].corr().iloc[0,1]**2
````
